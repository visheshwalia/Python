```python
#creating dummies of categorical data for training data
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname], prefix=colname, drop_first=True)
    df = pd.concat([df, col_dummies], axis=1)
    df.drop( colname, axis = 1, inplace = True )
    return df


for c_feature in cat_df.columns:
    cat_df[c_feature] = cat_df[c_feature].astype('category')
    cat_df = create_dummies(cat_df , c_feature )
    
for c_feature in cat_df_test.columns:
    cat_df_test[c_feature] = cat_df_test[c_feature].astype('category')
    cat_df_test = create_dummies(cat_df_test , c_feature )    
    
```
